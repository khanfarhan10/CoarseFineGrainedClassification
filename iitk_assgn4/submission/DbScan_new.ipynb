```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import cv2
import numpy as np
import matplotlib
```

```python
img= cv2.imread("00979.jpg") 
labimg = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
feature_image=np.reshape(labimg, [-1, 3])
rows, cols, chs = labimg.shape
db = DBSCAN(eps=1.0, min_samples=100)
db.fit(feature_image)
labels = db.labels_
```

```python
plt.figure(2)
plt.subplot(2, 1, 1)
plt.imshow(img)
plt.axis('off')
plt.subplot(2, 1, 2)
plt.imshow(np.reshape(labels, [rows, cols]))
plt.axis('off')
plt.show()
```
